The dataset is taken from Iris Flower Dataset at Kaggle. Link: https://www.kaggle.com/arshid/iris-flower-dataset


This dataset is used to do a multi-class classification (3 classes). It only contains 150 records. The attributes in the set are:

* Petal Length: the length of the petal - modified leaf which surround the reproductive parts of a flower.
* Petal Width: the width of the petal.
* Sepal Length: the length of the sepal - the outermost part of a flower which encloses its developing reproductive structures.
* Sepal Width: the width of the sepal.
* Species: the species of the flower (Iris setosa, Iris versicolor or Iris virginica)

For those of you who love to start learning data science and machine learning, this notebook will be a small kick start for you to dive in those world. :) Enjoy it!

First, let's import the csv file from Kaggle. I have downloaded the file and it is in my repo, so I will import it from my github repo. 

```
import pandas as pd
iris = pd.read_csv('https://raw.githubusercontent.com/vincentjunitio00/Iris-Flower/master/Iris.csv')
```

Let's see how is the dataset looks like.

```
iris
```

I notice that the dataset is somehow sorted by the "Species" column. Let's have a check is it really true or not.

```
iris[iris['Species']=='Iris-setosa']
```

```
iris[iris['Species']=='Iris-versicolor']
```

```
iris[iris['Species']=='Iris-virginica']
```

Yes it is! Now I am going to shuffle the dataset (to prevent a clustered species since it will affect the learning process by Machine Learning). 

Do we really need to shuffle the dataset?
My answer would be yes! Shuffling the dataset will reduce the variance of the data and be more generalized so the prediction will be more accurate.

```
from sklearn.utils import shuffle
iris = shuffle(iris)
```

Now let's see the shuffled dataset (you may ignore the index of the dataset)

```
iris
```

Next, let's see what is the type of each columns/ attributes.

```
iris.info()
```

Suppose that initially I do not know how many datas with Iris versicolor, Iris setosa and Iris virginica. I am going to count it now.

```
iris['Species'].value_counts()
```

Oh it's 50 datas of each species.

Since our task is to do a multi-class classification or we can say that we are going to predict whether this data belongs to Iris setosa, iris versicolor or iris virginica, I will change each value with an integer - 0 for Iris setosa, 1 for Iris versicolor, 2 for Iris virginica

```
pd.options.mode.chained_assignment = None
def changeValue(dataframe):
  dataframe['Species'][dataframe['Species']=='Iris-setosa'] = 0
  dataframe['Species'][dataframe['Species']=='Iris-versicolor'] = 1
  dataframe['Species'][dataframe['Species']=='Iris-virginica']= 2
```

```
changeValue(iris)
```

```
iris.reset_index(drop=True)
```

As you see, the value in "Species" attribute is now changed into a number ranging from 0 to 2.

It's training time!
But before that, I have to split the dataset into training set and test set. Feel free to tweak the length but please make sure that the number of the training set is larger the number of the test set. Since I use 0.94 for the length variable, I will have 141 datas for training set and 9 datas for test set - in this case the number of set that we are going to predict is 9 datas.

```
length = 0.94
split = int(length * len(iris))
train_set, test_set = iris[:split], iris[split:]
```

```
train_set = train_set.reset_index(drop = True)
train_set
```

```
test_set = test_set.reset_index(drop = True)
test_set
```

The "Species" attribute will be fed into y_train variable. The "Id" attribute will not give us any information about the species so I am going to drop these two attributes. (Do not be confused with the variable name, sometimes people say columns, sometimes people say attributes, sometimes people say features. I personally make columns are the same as attributes, while features are the attributes that we take into account for the learning proses)

```
drop_features = ["Id","Species"]
x_train_data = train_set.drop(drop_features,axis = 1)
```

```
x_train_data
```

I will save the "Species" attribute to y_train variable.

```
y_train = train_set["Species"]
```

y_train should be in int64, so let's change its type.

```
y_train = y_train.astype('int64')
```

Import the machine learning model -> Support Vector Classification. Check this link below to learn more about Support Vector Classification. 
https://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html

```
from sklearn.svm import SVC
```

```
svc = SVC()
# fit it to start training
svc.fit(x_train_data, y_train)
```

I will repeat the same process like what I have done with the training set to the test set (specifying the "Species" to y_true and drop unused features/ attributes to x_test_set). 

```
y_true = test_set['Species']
```

```
y_true = y_true.astype('int64')
```

The first column is the index column, while the second column is the "Species" column. Since we are using the second column only, I will convert it into a numpy array.

```
y_true
```

You can see what's the difference after I have converted it into numpy array.

```
y_true_numpy = y_true.to_numpy()
y_true_numpy
```

```
x_test_set = test_set.drop(["Id","Species"],axis = 1)
```

```
x_test_set
```

After training, it's time to predict! I am going to predict the x_test_set. Remember that the x_test_set consists of datas in which the models have not yet trained.

```
predict = svc.predict(x_test_set)
```

```
predict
```

I use confusion matrix to see how accurate my prediction is.

```
from sklearn.metrics import confusion_matrix
confusion_matrix(y_true_numpy, predict, labels = [0,1,2])
```

The confusion matrix above is 3x3. Suppose the matrix is in a Cartesian coordinate. The y axis is the true label, while the x axis is the predict label. The accurate prediction will have a value in the main diagonal of the confusion matrix (index = [0][0], [1][1], [2][2]). I set the parameter labels = [0,1,2] showing the order of the columns and rows in the matrix. First column and row for the 0 -> Iris-setosa. Second column and row for the 1 -> Iris-versicolor. Third column and row for the 2 -> Iris-virginica.
